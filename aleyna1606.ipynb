# Import Libraries

**Pandas ve çizim kütüphanelerini içe aktarın. Sklearn'ü daha sonra içe aktarabilirsiniz.**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Verileri Alın

**DATA klasöründeki mushrooms.csv dosyasını bir dataframe olarak okuma.**

```python
df = pd.read_csv('./mushrooms.csv')
df.head()
```

```python
df['class'].unique()
```

**Görev: info() ve describe() yöntemlerine gözatın.**

```python
df.info()
```

```python
df.describe().T
```

# Keşifsel Veri Analizi

Renklerin uyumu konusunda endişelenmeyin, sadece olayın ana fikri ile ilgilenin.

**Görev: class etiketinin dengeli olup olmadığını görmek için countplot grafiğini çizin.**

```python
sns.countplot(data=df, x=df['class']);
```

**Görev: Veri setinde her bir özelliğin kategoriktir. Bu kategorik özelliklerin sahip oldukları 'unique'(farklı) string değerlerin sayısını *feat_uni* isimli bir dataframe olarak oluşturun.**

```python
feat_uni = df.describe().transpose().reset_index().sort_values(by='unique', ascending=True)
feat_uni
```

**Görev: *feat_uni* veri çerçevesindeki 'index' sütununa göre 'unique' sütunun countplot grafiğini çizin.**

```python
plt.figure(figsize=(14,6),dpi=200)
sns.barplot(data=feat_uni, x='index', y='unique');
```

# Verilerin ayarlanması

**GÖREV: X özellik ve y etiket setini oluşturun.**

```python
X = df.drop('class', axis=1)
y = df['class']
```

**GÖREV: X'te eksik veri olup olmadığını kontrol edin.**

```python
X.isnull().sum()
```

**GÖREV: X 'teki kategorik özellikler için kukla değişkenler atıyarak X'i tamamen sayısallaştırın ve listeleyin.**

```python
X = pd.get_dummies(X, drop_first=True)
X.head()
```

## Train Test Split

**GÖREV: Verilerinizi eğitim ve test seti olarak bölmek için test_size=0.3 ve random_state=101 kullanın.**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

## Karar Ağacı Modeli Oluşturma ve Eğitimi

**GÖREV: Karar ağacı sınıflandırma modeli için kütüphanesinden DecisionTreeClassifier nesnesini import edin.**

```python
from sklearn.tree import DecisionTreeClassifier
```

**GÖREV: dtree adlı bir DecisionTreeClassifier() örneği oluşturun ve bunu eğitim verilerine uydurun.**

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

### Eğitilmiş Ağacı Görselleştirme
**GÖREV: eğitim setine uydurulan dtree modeline ait ağacı görselleştirin.**

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(17,8))
plot_tree(dtree,feature_names=X.columns,filled=True)
plt.title('Eğitim Ağacın Görseli')
```

##  Karar Ağacı Modelinin Tahminleri ve Değerlendirilmesi
**GÖREV: Test setinden tahminleri elde edin. Sınıflandırma raporu ve karışıklık matrisi oluşturun.**

```python
from sklearn.metrics import classification_report, confusion_matrix,ConfusionMatrixDisplay
```

```python
predictions = dtree.predict(X_test)
```

```python
print(classification_report(y_test, predictions))
```

```python
cm = confusion_matrix(y_test,predictions)
ConfusionMatrixDisplay(confusion_matrix = cm,
                       display_labels=dtree.classes_).plot()
```

**GÖREV: Modelin özellik önemlerini listeleyin.**

```python
dtree.feature_importances_
```

**GÖREV: En büyük öneme sahip özellik ismini yazdırın ve ne anlama geldiğini açıklamaya çalışın.**

```python
dtree.feature_importances_.argmax()
```

```python
X.columns[22]
```

AÇIKLAMA: Modelimiz, tek suşa dayalı olarak odor_n:No Odor'un en önemli özellik olduğunu düşünüyor ve oldukça etkileyici bir doğruluk öngörüyor.

**GÖREV: Bu en büyük öneme sahip özelliği ait olduğu kategorik özelliğin countplot grafiğini hue değeri 'class' olacak şekilde çizdirin.**

```python
sns.countplot(data=df, x='odor', hue='class');
```

## Rastgele Orman modelini eğitme

**GÖREV: rfc isimli bir RandomForestClassifier() örneğini oluşturun ve eğitim verilerine uydurun.**

```python
rfc = DecisionTreeClassifier()
```

```python
rfc.fit(X_train, y_train)
```

## Tahminler ve Değerlendirme

y_test değerlerini tahmin edelim ve modeli değerlendirelim.

**GÖREV: X_test için tahminleri elde edin.**

```python
predictions = rfc.predict(X_test)
```

**GÖREV: Tahmin sonuçlarına ait sınıflandırma raporu oluşturun.**

```python
print(classification_report(y_test, predictions))
```

**GÖREV: Tahmin sonuçlarına ait karışıklı matrisi oluşturun.**

**GÖREV: Rastgele orman için özellik önem değerlerini listeleyin.**

```python
rfc.feature_importances_
```

**GÖREV: Özellik önem değerlerini index'i orijinal veri setinin sütunları olan feats isimli bir dataframe oluşturun ve son 5 satırını listeleyin.**

```python
features = pd.DataFrame(index=X.columns, data=rfc.feature_importances_, columns=['Importance'])
features.head()
```

**GÖREV: feats dataframe'inin 0.015'ten büyük değerlerinin olduğu imp_feats veri çerçevesini oluşturun ve uzunluğunu bulun.**

```python
imp_feats = features[features['Importance'] > 0].sort_values('Importance', ascending=False)
```

```python
imp_feats
```

**GÖREV: Rastgele orman için özellik önem değerlerini sıralayarak barplot grafiğini çizdirin.**

```python
plt.figure(figsize=(14, 6))
sns.barplot(data=imp_feats, x=imp_feats.index, y='Importance')
```
